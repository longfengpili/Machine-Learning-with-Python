```python
CSV_COLUMNS = ['key','fare_amount','pickup_datetime','pickup_longitude',
               'pickup_latitude','dropoff_longitude','dropoff_latitude','passenger_count']
LABEL_COLUMN = 'fare_amount'
DEFAULTS = [['nokey'], [0.0], ['nokey'], [-74.0], [40.0], [-74.0], [40.7], [1.0]]

def decode_csv(value_column):
    columns = tf.decode_csv(value_column, record_defaults = DEFAULTS)
    print(columns)
    features = dict(zip(CSV_COLUMNS, columns))
    print(features)
    label = features.pop(LABEL_COLUMN)
    label = dict(([(LABEL_COLUMN,label)]))
    print(label)
    return features, label


file_list = tf.gfile.Glob('E:/new-york-city-taxi-fare-prediction/train.csv')
file_list
dataset_ = tf.data.TextLineDataset(file_list).skip(1)
# print(dataset_)
dataset_ = dataset_.map(decode_csv)
# print(dataset_)

num_epochs = None # indefinitely
batch_size = 512
dataset_ = dataset_.shuffle(buffer_size = 10 * batch_size)
# print(dataset_)
dataset_ = dataset_.repeat(num_epochs).batch(batch_size)
# print(dataset_)

iterator = dataset_.make_one_shot_iterator()
with tf.Session() as sess:
    for i in range(1):
        sess.run(dataset_.make_one_shot_iterator().get_next())
```
